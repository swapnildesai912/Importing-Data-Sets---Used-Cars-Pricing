```python
import numpy as np
import pandas as pd

# Load the data again as the execution state was cleared
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data'
df = pd.read_csv(url, header=None)

df.columns = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]

# replace "?" to NaN
df.replace("?", np.nan, inplace = True)
df.head(5)
```

```python
missing_data = df.isnull()
missing_data.head(5)
```

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")
```

**Calculating the mean value for the "normalized-losses" column**

```python
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
```

**Replacing "NaN" with mean value in "normalized-losses" column**

```python
df.replace({"normalized-losses": np.nan}, avg_norm_loss, inplace=True)
```

**Calculating the mean value for the "bore" column**

```python
avg_bore=df['bore'].astype('float').mean(axis=0)
print("Average of bore:", avg_bore)
```

**Replacing "NaN" with the mean value in the "bore" column**

```python
df.replace({'bore': np.nan}, avg_bore, inplace=True)
```

**Calculating the mean value for the "Stroke" column**

```python
avg_stroke = df["stroke"].astype("float").mean(axis = 0)
print("Average of stroke:", avg_stroke)
```

**Replacing "NaN" with the mean value in the "Stroke" column**

```python
df.replace({"stroke": np.nan}, avg_stroke, inplace=True)
```

**Calculating the mean value for the "Horsepower" column**

```python
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print("Average horsepower:", avg_horsepower)
```

**Replacing "NaN" with the mean value in the "Horsepower" column**

```python
df.replace({'horsepower': np.nan}, avg_horsepower, inplace=True)
```

**Calculating the mean value for the "peak-rpm" column**

```python
avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
print("Average peak rpm:", avg_peakrpm)
```

**Replacing "NaN" with the mean value in the "peak-rpm" column**

```python
df.replace({'peak-rpm': np.nan}, avg_peakrpm, inplace=True)
```

**Count Most type No.of Doors**

```python
df['num-of-doors'].value_counts()
```

```python
df.replace({'num-of-doors': np.nan}, "Four", inplace=True)
```

**drop all rows that do not have price data**

```python
df.dropna(subset=["price"], axis=0, inplace=True)
df.reset_index(drop=True, inplace=True)
```

**Converting data types to proper format**

```python
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

**Converting mpg to L/100km by mathematical operation**

```python
df['city-L/100km'] = 235/df["city-mpg"]

df.head()
```

```python
df["highway-mpg"] = 235/df["highway-mpg"]

df.rename(columns={'"highway-mpg"':'highway-L/100km'}, inplace=True)

df.head()
```

```python
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
```

**normalizing the column "height".**

```python
df['height'] = df['height']/df['height'].max()

df[["length","width","height"]].head()
```

**Converting Data type into Correct Type**

```python
df["horsepower"]=df["horsepower"].astype(int, copy=True)
```

**Plot the histogram of horsepower to see the distribution of horsepower.**

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["horsepower"])

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

## ** Binning Data In Pandas**

```python
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
group_names = ['Low', 'Medium', 'High']
```

```python
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head(20)
```

```python
df["horsepower-binned"].value_counts()
```

**Plot the distribution of each bin**

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```
