**Importing Pandas and numpy**



```python
import pandas as pd

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data'
df = pd.read_csv(url)

df.head()
```

**Utilize the Pandas method read_csv() to load the data into a dataframe.**

```python
print("The first 5 rows of the dataframe")
df.head(5)
```

**showing the first 5 rows using dataframe.head() method**

```python
print("The last 10 rows of the dataframe\n")
df.tail(10)
```

**showing the Last 10 rows using dataframe.tail() method**

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n", headers)
```

**Creating  list of "headers" that include all column names in order.**

```python
df.columns = headers
df.columns
```

**Replace headers and recheck our data frame**

```python
df.head(10)
```

**Showing the first 10 entries of the updated data frame and note that the headers are updated**

```python
df1=df.replace('?',np.nan)
```

**replacing the "?" symbol with nan so the dropna() can remove the missing values**

**For droping missing values along the column "price"**

```python
df=df1.dropna(subset=["price"], axis=0)
df.head(20)
```

**`axis=0` means that the contents along the entire row will be dropped wherever the entity 'price' is found to be NaN**
